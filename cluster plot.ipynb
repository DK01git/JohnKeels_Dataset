```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('redemption_data.csv')
df.head()
```

```python
df.shape
```

```python
# Assuming you have a dataset named 'data' with features
# X1, X2, ..., Xn, where n is the number of features
```

```python
# Create an instance of the KMeans class
kmeans = KMeans(n_clusters=3)
```

```python
dfk = df
```

```python
#create a seperate column to take only the day from the date column
dfk['day'] = pd.DatetimeIndex(df['Date']).day
```

```python
#create a seperate dataframe without Date column
dfk = dfk.drop(['Date'], axis=1)
```

```python
dfk.head()
```

```python
# Fit the data to the KMeans model
kmeans.fit(dfk)
```

```python
# Get the cluster labels for each data point
labels = kmeans.labels_
labels
```

```python
# Get the cluster centers
centers = kmeans.cluster_centers_
centers
```

```python
# Plot the data points with different colors based on their cluster labels
#add data labels to the plot from Target_Group column

plt.scatter(dfk['day'], dfk['Order_Amount'], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], marker='x', color='red')
plt.xlabel('Day')
plt.ylabel('Redemption Category')
plt.title('K-means Clustering')
plt.legend()
plt.show()

```
