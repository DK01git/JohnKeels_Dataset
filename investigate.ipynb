```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
dff = df
dff.head()
```

```python
#select all columns except date column and convert them into numerical columns
dff = dff.select_dtypes(include=[np.number])
```

```python
#convert Target-Group and Distance Rank columns into numerical columns
dff['Target-Group'] = dff['Target-Group'].astype('category').cat.codes
dff['Distance Rank'] = dff['Distance Rank'].astype('category').cat.codes
```

```python
dff.head()  
```

```python
dff.info()
```

```python
x = dff[['Distance Rank', 'Target-Group', 'Redemption Category']]
y = dff['Order Amount']
```

```python
#split the data into train and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
# Train the model
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#predictions
y_pred = regressor.predict(x_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
#rename Redemption Category column to Redemption_Category
dff = dff.rename(columns={'Redemption Category':'Redemption_Category'})
```

```python
dff.head()
```

```python
#provide a new row with new values and predict the order amount
new_row = {'Distance Rank': 1, 'Target-Group': 1, 'Redemption_Category': 3}

#predict the order amount for the new row
regressor.predict([list(new_row.values())])
```

```python
#evaluate the model's accuracy using cross validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(regressor, x, y, cv=5)
print(scores)
print(scores.mean())
```

```python
#plot the predicted values against the actual values
plt.scatter(y_test, y_pred, color='red')
plt.xlabel('Actual Order Amount')
plt.ylabel('Predicted Order Amount')
plt.title('Actual vs Predicted Order Amount')
plt.show()
```

```python
#plot the residuals
plt.scatter(y_pred, y_test-y_pred, color='red')
plt.xlabel('Predicted Order Amount')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
#plot the results using scatter plot with regression line
sns.regplot(x=y_test, y=y_pred, color='red')
plt.xlabel('Actual Order Amount')
plt.ylabel('Predicted Order Amount')
plt.title('Actual vs Predicted Order Amount')
plt.show()
```
